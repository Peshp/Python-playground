```python
import pandas as pd 

data = {
    'Name': ['Amy White', 'Jack Steward', 'Richard Laudardale', 'Sara Johnson'],
    'Age': [50, 53, 35, 43],
    'Working Experiance (Yrs)': [5, 4, 3, 2]
}

df = pd.DataFrame(data)
print(df)
```

```python
address = ['Sofia', 'York', 'Berlin', 'Amsterdam']
df['Address'] = address

display(df)
```

```python
#Calculate the average age

df['Age'].mean()
```

```python
#average working experience

df['Working Experiance (Yrs)'].mean()
```

```python
import numpy as np

df['Seniority'] = np.where(df['Age'] >= 50, 'Senior', 'Junior')
display(df)
```

```python
df.sort_values(by='Working Experiance (Yrs)', ascending=False)
```
